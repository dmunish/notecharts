```python
import random
from opensimplex import OpenSimplex

noise = OpenSimplex(random.randint(0, 1000))

data = []
val_min = float('inf')
val_max = float('-inf')

scale = 12
number = 40

for i in range(number + 1):
    for j in range(number + 1):
        for k in range(number + 1):
            value = noise.noise3(i / scale, j / scale, k / scale)
            val_min = min(val_min, value)
            val_max = max(val_max, value)
            z = value * 2 + 4
            data.append([i, j, k, float(z)])
```

```python
from notecharts import Chart, Palette

options = {
    "textStyle": {
        "fontFamily": "Josefin Sans"
    },
    "title": {
        "text": "Simplex",
        "left": "center",
        "top": "3%",
        "textStyle": {
            "fontSize": 22,
            "fontWeight": 600,
        }
    },
    "toolbox": {
        "feature": {
            "saveAsImage": {"pixelRatio": 3}
        }
    },
    "visualMap": {
        "show": False,
        "min": 2,
        "max": 6,
        "inRange": {
            "symbolSize": [0.1, 8],
            "color": Palette(palette="Plasma", n=8),
            "colorAlpha": [0.2, 1]
        }
    },
    "xAxis3D": {"type": "value"},
    "yAxis3D": {"type": "value"},
    "zAxis3D": {"type": "value"},
    "grid3D": {
        "viewControl": {
            "projection": "orthographic",
            "autoRotate": True,
            "orthographicSize": 200,
            "autoRotateSpeed": 20,
            "damping": 0.9,
            "alpha": 20,
            "beta": 45
        },
        "axisLabel":{
            "textStyle":{
                "color": "white"
            }
        },
        "axisLine":{
            "lineStyle":{
                "color": "white"
            }
        }
    },
    "series": [{
        "type": "scatter3D",
        "data": data,
        "symbolSize": 5
    }]
}

Chart(options, width="650px", theme="dark").display()
```
