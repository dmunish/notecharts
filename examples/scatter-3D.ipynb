```python
from opensimplex import OpenSimplex

noise = OpenSimplex(23)

data = []
val_min = float('inf')
val_max = float('-inf')

scale = 15
number = 40

for i in range(number + 1):
    for j in range(number + 1):
        for k in range(number + 1):
            value = noise.noise3(i / scale, j / scale, k / scale)
            val_min = min(val_min, value)
            val_max = max(val_max, value)
            z = value * 2 + 4
            data.append([i, j, k, float(z)])
```

```python
from notecharts import Chart, Palette, Option

options: Option = {
    "visualMap": {
        "calculable": True,
        "realtime": False,
        "min": 2,
        "max": 6,
        "inRange": {
            "symbolSize": [0.1, 10],
            "color": Palette(palette="Matter", saturation="+0.05", value="+0.1", n=8, reverse=True),
            # "colorAlpha": [0.3, 0.9]
        }
    },
    "toolbox": {
        "feature": {
            "saveAsImage": {}
        }
    },
    "xAxis3D": {"type": "value", "nameTextStyle": { "color": "transparent"}},
    "yAxis3D": {"type": "value", "nameTextStyle": { "color": "transparent"}},
    "zAxis3D": {"type": "value", "nameTextStyle": { "color": "transparent"}},
    "grid3D": {
        "viewControl": {
            "projection": "orthographic",
            "orthographicSize": 200,
            "damping": 0.9,
            "alpha": 30,
            "beta": 45
        },
        "axisLine":{
            "lineStyle":{
                "color": "white"
            }
        },
        "axisLabel": {"show": False},
        "axisTick": {"show": False},
    },
    "series": [{
        "type": "scatter3D",
        "data": data,
        "emphasis": {
            "label": {"show": False}
        },
    }]
}

Chart(options, width="700px", theme="dark", devicePixelRatio=2).display()
```
