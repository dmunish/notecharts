```python
import random
from notecharts import Chart, Palette, Option

series_names = ["Primary", "Secondary", "Tertiary", "Quaternary"]
indicator_names = ["Equipment", "Materials", "Food & Beverage", "Apparel", "Tourism", "Technology"]

data_values = [[random.randint(5000, 25000) for _ in indicator_names] for _ in series_names]
max_values = [30000] * len(indicator_names) 

colors = Palette("SunsetDark", len(series_names))

options: Option = {
    "textStyle": {"fontFamily": "Poppins"},
    "tooltip": {
        "trigger": "item",
        "borderWidth": 0,
    },
    "legend": {
        "icon": "circle"
    },
    "radar": {
        "indicator": [{"name": n, "max": m} for n, m in zip(indicator_names, max_values)],
        "shape": "polygon"
    },
    "series": [
        {
            "name": name,
            "type": "radar",
            "symbol": "circle",
            "symbolSize": 8,
            "itemStyle": {
                "color": color, 
            },
            "lineStyle": {
                "width": 2, 
                "shadowBlur": 8, 
                "shadowColor": color
            },
            "areaStyle": {
                "opacity": 0.05
            },
            "data": [{"value": val, "name": name}],
            "emphasis": {
                "focus": "series",
                "areaStyle": {
                    "opacity": 0.5
                },
            }
        }
        for name, color, val in zip(series_names, colors, data_values)
    ]
}

Chart(options, theme="dark", width="700px").display()
```

```python
import pandas as pd
import random
from notecharts import Radar
 
random.seed()
 
df = pd.DataFrame({
    "Category": ["Primary", "Secondary", "Tertiary", "Quaternary"],
    "Equipment": [random.randint(4000, 6000) for _ in range(4)],
    "Materials": [random.randint(10000, 16000) for _ in range(4)],
    "Food & Beverage": [random.randint(10000, 30000) for _ in range(4)],
    "Apparel": [random.randint(10000, 40000) for _ in range(4)],
    "Tourism": [random.randint(10000, 50000) for _ in range(4)],
    "Technology": [random.randint(5000, 15000) for _ in range(4)],
})
 
Radar(
    data=df,
    series_col="Category",
    title="Radar",
    theme="dark",
    palette="agGrnYl",
    width="600px",
    font="Poppins",
).display()
```
