```python
from notecharts import Chart, Palette, Option
import random

random.seed()

series_names = ["Primary", "Secondary", "Tertiary", "Quaternary"]
indicator_names = [
    "Equipment", "Materials", "Food & Beverage",
    "Apparel", "Tourism", "Technology",
]
ranges = [
    (4000, 6000), (10000, 16000), (10000, 30000),
    (10000, 40000), (10000, 50000), (5000, 15000),
]

cols = [[random.randint(lo, hi) for _ in series_names] for lo, hi in ranges]
data_values = [[col[i] for col in cols] for i in range(len(series_names))]
max_values = [round(max(col) * 1.1) if max(col) > 0 else 100 for col in cols]

n = len(series_names)
color_base = Palette("SunsetDark", n)
color_trans = Palette("SunsetDark", n, alpha=0, value="-0.5")

series_list = []
for i in range(n):
    base_c = color_base[i]
    trans_c = color_trans[i]
    series_list.append({
        "name": series_names[i],
        "type": "radar",
        "symbol": "circle",
        "symbolSize": 10,
        "itemStyle": {
            "opacity": 0,
            "color": base_c,
            "shadowBlur": 10,
            "shadowColor": base_c,
        },
        "lineStyle": {"color": base_c, "width": 2},
        "areaStyle": {
            "opacity": 0.3,
            "color": {
                "type": "linear",
                "x": 0, "y": 0, "x2": 1, "y2": 1,
                "colorStops": [
                    {"offset": 0, "color": base_c},
                    {"offset": 1, "color": trans_c},
                ],
            },
        },
        "data": [{"value": data_values[i], "name": series_names[i]}],
        "emphasis": {
            "focus": "series",
            "lineStyle": {"width": 3, "opacity": 1},
            "itemStyle": {"opacity": 1},
            "areaStyle": {"opacity": 1.0},
        },
        "blur": {
            "lineStyle": {"width": 1, "opacity": 0.3},
            "itemStyle": {"opacity": 0},
            "areaStyle": {"opacity": 0.2},
        },
    })

options: Option = {
    "title": {
        "text": "Radar",
        "left": "center",
        "top": 24,
        "textStyle": {"fontSize": 22, "fontWeight": 600},
    },
    "textStyle": {"fontFamily": "Poppins"},
    "tooltip": {
        "trigger": "item",
        "borderWidth": 0,
        "padding": [14, 18],
        "shadowBlur": 10,
        "shadowColor": "rgba(0, 0, 0, 0.12)",
        "extraCssText": "border-radius: 12px;"
    },
    "legend": {
        "data": series_names,
        "bottom": 20,
        "icon": "circle",
        "itemWidth": 20,
        "itemGap": 30,
        "itemStyle": {"opacity": 1},
    },
    "toolbox": {
        "feature": {
            "restore": {},
            "saveAsImage": {"pixelRatio": 3},
        },
    },
    "radar": {
        "radius": "60%",
        "center": ["50%", "50%"],
        "startAngle": 90,
        "triggerEvent": True,
        "shape": "polygon",
        "indicator": [
            {"name": indicator_names[j], "max": max_values[j]}
            for j in range(len(indicator_names))
        ],
    },
    "animationEasing": "quinticOut",
    "animationDuration": 900,
    "series": series_list,
}

Chart(options, theme="dark").display()
```
